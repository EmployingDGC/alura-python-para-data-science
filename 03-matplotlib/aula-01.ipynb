```python
import matplotlib.pyplot as plt
import random as rd
```

```python
qtt_grades = 8
```

```python
grades = [rd.randint(0, 10) for _ in range(qtt_grades)]
```

```python
axis_x = tuple(range(1, qtt_grades + 1))
```

```python
plt.plot(axis_x, grades, marker="o")
plt.title("Notas de matemática")
plt.xlabel("Provas")
plt.ylabel("Notas")
plt.show()
```
