```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc("figure", figsize=(20, 10))
```

```python
df = pd.read_csv("./aluguel.csv", sep=";")
```

```python
df
```

```python
df.groupby("Bairro").Valor.mean()
```

```python
df.groupby("Bairro")[["Valor", "Condominio"]].mean().round(2)
```

```python
df.groupby("Bairro").Valor.describe().round(2)
```

```python
df.groupby("Bairro").Valor.agg(["min", "max", "sum"])
```

```python
bar = df.groupby("Bairro").Valor.mean().plot.bar(color="blue")
bar.set_ylabel("Valor do Aluguel")
bar.set_xlabel("Bairro")
bar.set_title("Valor Médio do Aluguel por Bairro", {
    "fontsize": 22
})
```
