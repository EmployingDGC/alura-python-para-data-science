```python
import pandas as pd
```

```python
df = pd.read_csv("./aluguel.csv", sep=";")
```

```python
df
```

```python
steps = [-1, 0, 2, 4, 6, 8, 10, 12, df.Quartos.max()]
```

```python
df.assign(GrupoQuartos=pd.cut(df.Quartos, steps))
```

```python
lables = [
    "0 Quartos",
    "1 e 2 Quartos",
    "3 e 4 Quartos",
    "5 e 6 Quartos",
    "7 e 8 Quartos",
    "9 e 10 Quartos",
    "11 e 12 Quartos",
    "13 Quartos ou Mais"
]
```

```python
pd.cut(df.Quartos, steps, labels=lables).value_counts().sort_index()
```
